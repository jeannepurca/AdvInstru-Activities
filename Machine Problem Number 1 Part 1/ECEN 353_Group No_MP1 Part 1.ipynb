Now, it can be seen that the Capital Gains column has been dropped from the dataframe.

<h3>Correct Data Format</h3>
<p>The last step in data cleaning is checking and making sure that all data is in its correct format (int, float, text, etc.).</p>
<p>Now, it can be seen that there is a column for dates, which also includes timezone information. Since we're working with stock data and aiming to predict the "Close" price, time is a crucial factor. However, the raw date string isn't directly useful in its current form.</p>
<p>With this, the format of the <b>'Date'</b> column must be changed.</p>

```python
# Convert the "Date" column to a datetime format.
dataset['Date'] = pd.to_datetime(dataset['Date'], errors='coerce', utc=True)
```

```python
# Extract useful features.
dataset['Year'] = dataset['Date'].dt.year
dataset['Month'] = dataset['Date'].dt.month
dataset['Day'] = dataset['Date'].dt.day
dataset['DayOfWeek'] = dataset['Date'].dt.dayofweek
dataset['IsWeekend'] = dataset['DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)
dataset['Days_Since_Start'] = (dataset['Date'] - dataset['Date'].min()).dt.days

# Drop the original Date column.
dataset.drop('Date', axis=1, inplace=True)

# Display the corrected dataframe.
dataset.head(40)
```

Since several features were extracted from the "Date" column, it's important to recheck the dataset for any missing values that may have been introduced during the process.

```python
# Check the rows of data in each column
dataset_count = dataset.count()
dataset_count
```

The <b>"Close"</b> column will be the one predicted, it can be moved to the rightest side of the dataframe for the table to look more organized.

```python
# Move "Close" column to the end
target = dataset.pop('Close')  
dataset['Close'] = target     
dataset.head()
```

<h3>Correlation and Causation</h3>
<p><b>Correlation</b>: a measure of the extent of interdependence between variables.</p>
<p><b>Causation</b>: the relationship between cause and effect between two variables.</p>

<p>Determining correlation is much simpler than determining causation as causation may require independent experimentation.</p>

<h4>Pearson Correlation</h4>
<p>The Pearson Correlation measures the linear dependence between two variables X and Y.</p>
<p>The resulting coefficient is a value between -1 and 1 inclusive, where:</p>
<ul>
    <li><b>1</b>: Total positive linear correlation.</li>
    <li><b>0</b>: No linear correlation, the two variables most likely do not affect each other.</li>
    <li><b>-1</b>: Total negative linear correlation.</li>
</ul>
<p>Pearson Correlation is the default method of the function "corr".  Like before we can calculate the Pearson Correlation of the of the 'int64' or 'float64'  variables.</p>

```python
# To generate the Pearson Correlations Coefficient between variables
clean_dataset.corr
```

 Sometimes we would like to know the significant of the correlation estimate. 

<h4>P-value</h4>
<p>The P-value is the probability value that the correlation between these two variables is statistically significant. Normally, we choose a significance level of 0.05, which means that we are 95% confident that the correlation between the variables is significant.</p>

By convention, when
<ul>
    <li>the p-value is $<$ 0.001: we say there is strong evidence that the correlation is significant.</li>
    <li>the p-value is $<$ 0.05: there is moderate evidence that the correlation is significant.</li>
    <li>the p-value is $<$ 0.1: there is weak evidence that the correlation is significant.</li>
    <li>the p-value is $>$ 0.1: there is no evidence that the correlation is significant.</li>
</ul>

 We can obtain this information using  "stats" module in the "scipy"  library.

```python
# To import Stats Library from Scipy
from scipy import stats
```

Let us calculate the <b>Pearson Correlation Coefficient</b> and <b>P-value</b> of the variables:

```python
# To calculate the Pearson Correlation Coefficient and the P-Value between "Open" and "Close".
r, p = stats.pearsonr(clean_dataset["Open"], clean_dataset["Close"])

# To display the results
print("The Pearson Coefficient of Correlation is", r, "with a p-value of", p)
```

<b>Conclusion:</b> <i>Since the p-value is $<$ 0.001, the correlation between Open and Close is statistically significant, and the linear relationship is extremely strong (~0.999, can be rounded off to 1)</i></p>
