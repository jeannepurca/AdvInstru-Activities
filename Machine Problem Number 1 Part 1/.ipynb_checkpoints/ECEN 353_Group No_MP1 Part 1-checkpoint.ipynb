<h3>Import the libraries.</h3>

```python
# Import the Libraries (Pandas, Matplotlib, numpy)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

<h3>Import and display the dataset.</h3>

Use the Pandas method <b>read_csv()</b> to load the data. Then, use the method <b>head()</b> to display the first ten rows of the dataset.

```python
# Import the Dataset.
dataset = pd.read_csv("Students_Grading_Dataset.csv")

# Display the first ten rows of the dataset
dataset.head(10)
```

<h3>Identify the missing values.</h3>

Based on what was displayed, several <b>NaN</b> already appeared in the dataframe, which means there are missing values that may hinder the analysis. However, based on the <b><i>csv</i></b> file of the dataset, some of these NaN are from the features which include <b><i>"None"</i></b> as their values, while others are actual missing values.

With this, the values must be corrected.

To do retain the "None" data, the function <b>keep_default_na=False</b> must be inserted to the declaration of the csv file.

```python
# Retain the "None" data by inserting keep_default_na=False
dataset = pd.read_csv("Students_Grading_Dataset.csv", keep_default_na=False)
```

The actual missing data comes with a blank cell. With this, replace " " or "" with NaN. Use the function <b>.replace(A, B, inplace = True)</b>

```python
# Replace " " with NaN
dataset.replace(["", " "], np.nan, inplace=True)
```

```python
# Display the first ten rows of the dataset
dataset.head(10)
```

```python
# Check the rows of data in each column
dataset_count = dataset.count()
dataset_count
```

```python
# Check if there are missing data.
missing_data = dataset.isnull().sum().sort_values(ascending=False)
missing_data.head(10)
```
